## Importing Data

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Fitness_trackers_updated.csv')
df
```

# Central Tendency

```python
df.info()
```

Based on this, we need to standardize the variable by changing the variable to their used dtypes, for example Selling Price and Original Price.

```python
# Convert Selling Price and Original Price to float

## Convert Selling Price to Float
df["Selling Price"] = df["Selling Price"].str.replace(",", "")
df["Selling Price"] = df["Selling Price"].astype(float)

## Convert Original Price to Float
df["Original Price"] = df["Original Price"].str.replace(",", "")
df["Original Price"] = df["Original Price"].astype(float)

## Print the column Selling Price and Original Price
df[['Selling Price', 'Original Price']]
```

```python
## Check the data type
df.info()
```

With this variables, we can check the central tendency of the selling price. Code below will show the mean, median, and mode of the selling price.

```python
# Central Tendency of Selling Price
data = {
    'Mean': df['Selling Price'].mean(),
    'Mode': df['Selling Price'].mode().values[[0]],
    'Median': df['Selling Price'].median(),
    'Standard Deviation': df['Selling Price'].std(),
    'Variance': df['Selling Price'].var(),
    'Minimum': df['Selling Price'].min(),
    'Maximum': df['Selling Price'].max()
}

statistics_df = pd.DataFrame(data)

statistics_df
```

# Detect and Handle Outlier

```python
# Import library to draw boxplot
import seaborn as sns

# Drawing Boxplot for all numeric variable
plt.figure(figsize=(10, 5))
sns.boxplot(data=df[['Selling Price', 'Original Price']])
plt.show()
```

Based on these boxplot, there are so many outliers in the data. Now we will find out how many outliers in the data for Selling Price and Original Price.

```python
# Count outlier in the Selling Price

## Get the 1st and 3rd quartiles
Q1, Q3 = df['Selling Price'].quantile([0.25, 0.75])

## Calculate the interquartile range
IQR = Q3 - Q1

## Define the upper and lower bounds for outliers
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

## Define outlier
outliers = df.loc[(df['Selling Price'] <= lower_bound) | (df['Selling Price'] >= upper_bound)]

## Print count of outlier
print("Outlier in the Selling Price", outliers.shape[0])
```

```python
# Count outlier in the Original Price

## Get the 1st and 3rd quartiles
Q1, Q3 = df['Original Price'].quantile([0.25, 0.75])

## Calculate the interquartile range
IQR = Q3 - Q1

## Define the upper and lower bounds for outliers
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

## Define outlier
outliers = df.loc[(df['Original Price'] <= lower_bound) | (df['Original Price'] >= upper_bound)]

## Print count of outlier
print("Outlier in the Original Price", outliers.shape[0])
```

Based on the output above, the Selling Price variable has 37 outliers while the Original Price variable has 35 outlier. We want to handle this by dropping those values. We will choose to drop the outlier from the Selling Price variable.

```python
# Count outlier in the Selling Price

## Get the 1st and 3rd quartiles
Q1, Q3 = df['Selling Price'].quantile([0.25, 0.75])

## Calculate the interquartile range
IQR = Q3 - Q1

## Define the upper and lower bounds for outliers
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

## Define outlier
outliers = df.loc[(df['Selling Price'] <= lower_bound) | (df['Selling Price'] >= upper_bound)]

## Print count of outlier
df = df.drop(outliers.index, axis=0)
```

```python
# Draw Boxplot for Selling Price and Original Price

## Define the size of the figure 10 by 5.
plt.figure(figsize=(10, 5))

## Create boxplot from seaborn with variable Selling Price and Original Price
sns.boxplot(data=df[['Selling Price', 'Original Price']])

## Display the plot
plt.show()
```

Based on the output above, we can see that the outliers has been dropped from the data. Now we can check the boxplot again to see the difference. The new boxplot show that there are new outliers. We will count the number of outlier in the selling price.

```python
# Count outlier in the Original Price

## Get the 1st and 3rd quartiles
Q1, Q3 = df['Original Price'].quantile([0.25, 0.75])

## Calculate the interquartile range
IQR = Q3 - Q1

## Define the upper and lower bounds for outliers
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

## Define outlier
outliers = df.loc[(df['Original Price'] <= lower_bound) | (df['Original Price'] >= upper_bound)]

## Print count of outlier
print("Outlier in the Original Price", outliers.shape[0])
```

The result of outlier handling earlier drop the outlier by 12 from the previous data which is already good. We can keep dropping the outlier to get better data, but we will stop here as it will risk the representative of the data.

# Draw Bar Chart for Top 10 Brands

```python
# Count Plot for top ten sold brand
countsbar = df["Brand Name"].value_counts()[:10]
print("Count untuk Tiap Brand")
print(countsbar)

## Get labels and values
labels = countsbar.index.tolist()
print("Label")
print(labels)

values = countsbar.tolist()
print("Values")
print(values)

## Draw bar plot
fig, ax = plt.subplots()
ax.barh(labels, values)
ax.set_title("Brand Counts")
plt.show()
```

# Pie Chart for Top 5 Brands

```python
# Pie Chart for Top 5 Brands
## Get Value of total Brand Counts
countspie = df["Brand Name"].value_counts()[:5]
print("Count untuk Tiap Brand")
print(countspie)

## Get labels and values
labels = countspie.index.tolist()
values = countspie.tolist()
print("Labels dan Value")
print(labels)
print(values)

## Add other values in pie chart
total_count = df["Brand Name"].count()
other_count = total_count - sum(values)
print("Jumlah")
print(total_count)
print("Other count")
print(other_count)

## Add other values in labels and values
labels.append("Lainnya")
print(labels)
values.append(other_count)
print(values)

## Draw pie chart
fig, ax = plt.subplots()
ax.pie(values, labels=labels, autopct='%1.1f%%')
ax.set_title("Top 5 Brands")
plt.show()
```

# SQL Database
Source: https://www.kaggle.com/datasets/kaggle/sf-salaries

```python
# Import SQLITE database
import sqlite3

## Import SQLITE data 
conn = sqlite3.connect('database.sqlite')

cursor = conn.cursor()


cursor.execute('SELECT JobTitle, AVG(BasePay) AS avg_base_pay FROM Salaries GROUP BY JobTitle ORDER BY AVG(BasePay) DESC')

job_titles = []
avg_base_pay_values = []

rows = cursor.fetchall()

for row in rows:
    job_titles.append(row[0])
    avg_base_pay_values.append(row[1])

cursor.close()
conn.close()

# Create a bar chart using Matplotlib
plt.figure(figsize=(12, 6))
bars = plt.barh(job_titles[:10], avg_base_pay_values[:10])
plt.xlabel('Average Base Pay')
plt.ylabel('Job Title')
plt.title('Top 10 Job Titles by Average Base Pay')
plt.tight_layout()

for i, bar in enumerate(bars):
    value = round(avg_base_pay_values[i], 2)
    plt.text(bar.get_width() - 0.05, bar.get_y() + bar.get_height() / 2, value, ha='right', va='center', color='white')

# Display the chart
plt.show()
```
